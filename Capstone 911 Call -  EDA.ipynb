```python
df['Reason']=df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
plt.figure(figsize=(6,8))
sns.countplot(df['Reason'],data=df,palette='Set2')
```

```python
df['timeStamp']=pd.to_datetime(df['timeStamp'])
```

```python
df['timeStamp']
```

```python
time=df['timeStamp'].iloc[0]
```

```python
time.hour
```

```python
df['Hour']=df['timeStamp'].apply(lambda time:time.hour)
```

```python
df['Month']=df['timeStamp'].apply(lambda time:time.month)
```

```python
df['Day of Week']=df['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
dmap={0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week']=df['Day of Week'].map(dmap)
```

```python
df['Day of Week']
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=df['Day of Week'],data=df,hue=df['Reason'])
```

```python
plt.figure(figsize=(12,8))
sns.countplot(df['Month'],data=df,hue=df['Reason'])
```

```python
byMonth=df.groupby('Month').count()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(20,10))
byMonth['lat'].plot(linewidth=3,marker='o')
```

```python
plt.figure(figsize=(20,10))
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index(),size=10)
```

```python
df['Date']=df['timeStamp'].apply(lambda time:time.date())
```

```python
df.head()
```

```python
sns.set_style('white')
plt.figure(figsize=(15,8))
df.groupby('Date').count()['lat'].plot(linewidth=2)
plt.tight_layout()
```

```python
sns.set_style('white')
plt.figure(figsize=(20,10))
df[df['Reason']=='Fire'].groupby('Date').count()['lat'].plot(linewidth=3)
plt.tight_layout()
```
